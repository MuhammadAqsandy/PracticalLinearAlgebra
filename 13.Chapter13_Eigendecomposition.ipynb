## Computing Eigenvalues and Eigenvectors

```python
A = np.array([[4, 2],[1, 3]], dtype=float)
eigenvalues, eigenvectors = np.linalg.eig(A)
print('Eigenvalues:', eigenvalues)
print('Eigenvectors (columns):\n', np.round(eigenvectors, 4))
# Verify: A @ v = λ * v
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]; v = eigenvectors[:,i]
    print(f'λ={lam:.2f}: A@v={np.round(A@v,4)}, λ*v={np.round(lam*v,4)}, match={np.allclose(A@v, lam*v)}')
```

## Diagonalization: A = V D V^-1

```python
A = np.array([[4,2],[1,3]], dtype=float)
V, d = np.linalg.eig(A)
# V = eigenvector matrix, D = diagonal eigenvalue matrix
D = np.diag(d)
# Reconstruct A
A_recon = V @ D @ np.linalg.inv(V)
print('Original A:\n', A)
print('Reconstructed A:\n', np.round(A_recon, 6))
```

## Symmetric Matrix (Real Eigenvalues)

```python
S = np.array([[4,2,0],[2,3,1],[0,1,2]], dtype=float)
eigvals, eigvecs = np.linalg.eigh(S)  # for symmetric
print('Eigenvalues (real):', np.round(eigvals, 4))
print('Eigenvectors orthogonal:', np.allclose(eigvecs.T @ eigvecs, np.eye(3)))
```

## Power Iteration (Dominant Eigenvalue)

```python
def power_iteration(A, num_iter=100):
    b = np.random.rand(A.shape[0])
    for _ in range(num_iter):
        b_new = A @ b
        b = b_new / np.linalg.norm(b_new)
    eigenvalue = (A @ b) @ b / (b @ b)
    return eigenvalue, b
A = np.array([[2,1],[1,3]], dtype=float)
lam, v = power_iteration(A)
print('Dominant eigenvalue:', round(lam, 6))
print('NumPy check:', round(max(np.linalg.eigvals(A)), 6))
```

## Eigenvalues and Matrix Properties

```python
A = np.array([[3,1],[1,3]], dtype=float)
eigvals = np.linalg.eigvals(A)
print('Eigenvalues:', eigvals)
print('Trace = sum of eigenvalues:', np.trace(A), '=', sum(eigvals))
print('Det = product of eigenvalues:', round(np.linalg.det(A),4), '=', round(np.prod(eigvals),4))
```

## 📝 Summary

Chapter 13 introduces eigendecomposition — one of the most important concepts in linear algebra. An eigenvector v of matrix A is a special vector that only scales (not rotates) when multiplied by A: Av = λv, where λ is the eigenvalue. Diagonalization A = VDV⁻¹ expresses a matrix as a product involving its eigenvectors, making powers and exponentials of matrices efficient. Symmetric matrices always have real eigenvalues and orthogonal eigenvectors (spectral theorem), making them especially useful. Eigenvalues encode key matrix properties: trace = Σλᵢ, determinant = Πλᵢ.
