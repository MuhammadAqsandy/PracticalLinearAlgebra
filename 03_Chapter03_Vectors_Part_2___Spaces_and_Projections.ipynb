## Linear Combinations

```python
import numpy as np
v1 = np.array([1, 0])
v2 = np.array([0, 1])
# Linear combination: 3*v1 + 2*v2
result = 3*v1 + 2*v2
print('3*v1 + 2*v2 =', result)
```

## Checking Linear Independence

```python
# Two vectors are linearly independent if neither is a scalar multiple of the other
v1 = np.array([1, 2])
v2 = np.array([2, 4])  # 2 * v1 → linearly DEPENDENT
v3 = np.array([1, 3])  # not multiple of v1 → linearly INDEPENDENT
A = np.array([v1, v2])
rank_dep = np.linalg.matrix_rank(A)
print('Rank (v1, v2 - dependent):', rank_dep)  # 1
A2 = np.array([v1, v3])
print('Rank (v1, v3 - independent):', np.linalg.matrix_rank(A2))  # 2
```

## Orthogonality - Dot Product = 0

```python
a = np.array([1, 0])
b = np.array([0, 1])
print('Dot product of orthogonal vectors:', np.dot(a, b))  # 0
angle = np.degrees(np.arccos(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))))
print('Angle between vectors:', angle, 'degrees')
```

## Vector Projection

```python
# Project vector b onto vector a
a = np.array([3.0, 1.0])
b = np.array([2.0, 3.0])
# Scalar projection
scalar_proj = np.dot(b, a) / np.linalg.norm(a)
# Vector projection
vec_proj = (np.dot(b, a) / np.dot(a, a)) * a
print('Scalar projection of b onto a:', scalar_proj)
print('Vector projection:', vec_proj)
```

## Cosine Similarity

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
cosine_sim = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
print('Cosine similarity:', cosine_sim)
print('Angle:', np.degrees(np.arccos(cosine_sim)), 'degrees')
```

## 📝 Summary

Chapter 3 dives deeper into vector spaces. Linear combinations form new vectors from scaled sums of existing ones. Linear independence means no vector in a set can be expressed as a combination of the others — this is critical for defining basis sets. Orthogonal vectors have zero dot product (perpendicular), a property heavily used in PCA and QR decomposition. Vector projection decomposes one vector into components along another, which underpins least squares regression.
