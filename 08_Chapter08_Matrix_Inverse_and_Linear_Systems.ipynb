## Matrix Inverse

```python
import numpy as np
A = np.array([[1, 2],[3, 4]], dtype=float)
A_inv = np.linalg.inv(A)
print('A:\n', A)
print('A_inv:\n', np.round(A_inv, 4))
print('A @ A_inv (should be I):\n', np.round(A @ A_inv, 10))
```

## When is a Matrix Invertible?

```python
# Invertible: det != 0, full rank
A = np.array([[1, 2],[3, 4]])
B = np.array([[1, 2],[2, 4]])  # Singular!
print('det(A):', np.linalg.det(A))  # -2
print('det(B):', np.linalg.det(B))  # 0
try:
    np.linalg.inv(B)
except np.linalg.LinAlgError as e:
    print('Error inverting B:', e)
```

## Solving Ax = b with Inverse

```python
A = np.array([[2, 1],[1, 3]], dtype=float)
b = np.array([5, 10], dtype=float)
# Method 1: x = A_inv @ b (less stable)
x1 = np.linalg.inv(A) @ b
# Method 2: np.linalg.solve (recommended)
x2 = np.linalg.solve(A, b)
print('Via inverse:', np.round(x1, 6))
print('Via solve:', np.round(x2, 6))
```

## Moore-Penrose Pseudoinverse

```python
# For non-square or singular matrices
A = np.array([[1, 2],[3, 4],[5, 6]], dtype=float)  # 3x2, not square
A_pinv = np.linalg.pinv(A)
print('A shape:', A.shape)
print('Pseudoinverse shape:', A_pinv.shape)
print('A_pinv @ A (≈ I):\n', np.round(A_pinv @ A, 6))
```

## Condition Number - Numerical Stability

```python
A_good = np.array([[1, 0],[0, 1]], dtype=float)
A_bad  = np.array([[1, 1],[1, 1.0001]], dtype=float)
print('Condition number (good):', np.linalg.cond(A_good))
print('Condition number (bad):', round(np.linalg.cond(A_bad), 2))
print('High condition number = ill-conditioned = numerically unstable')
```

## 📝 Summary

Chapter 8 covers matrix inversion and solving linear systems. A matrix inverse A⁻¹ exists only when the determinant is non-zero (full rank). Using np.linalg.solve() is preferred over computing the inverse explicitly for numerical stability. The Moore-Penrose pseudoinverse extends the concept to non-square and singular matrices, forming the basis of the least-squares solution. The condition number measures numerical stability — a high condition number means small input changes cause large output changes (ill-conditioned system).
