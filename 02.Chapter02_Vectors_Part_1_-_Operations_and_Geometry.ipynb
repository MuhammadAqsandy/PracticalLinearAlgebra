## Vector Addition & Subtraction

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print('a + b =', a + b)
print('a - b =', a - b)
```

## Scalar Multiplication

```python
v = np.array([1, 2, 3])
scalar = 3
print('3 * v =', scalar * v)
print('-1 * v =', -1 * v)
```

## Dot Product

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot = np.dot(a, b)
dot_alt = a @ b
print('Dot product (np.dot):', dot)
print('Dot product (@):', dot_alt)
# Manual: 1*4 + 2*5 + 3*6
print('Manual:', 1*4 + 2*5 + 3*6)
```

## Vector Norm

```python
v = np.array([3, 4])
norm_l2 = np.linalg.norm(v)          # L2 norm
norm_l1 = np.linalg.norm(v, ord=1)  # L1 norm
print('L2 Norm (Euclidean):', norm_l2)  # should be 5.0
print('L1 Norm (Manhattan):', norm_l1)
```

## Unit Vector

```python
v = np.array([3.0, 4.0])
unit_v = v / np.linalg.norm(v)
print('Unit vector:', unit_v)
print('Norm of unit vector:', np.linalg.norm(unit_v))
```

## Geometric Visualization

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
# Addition
a = np.array([2, 1]); b = np.array([1, 3])
axes[0].quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,color='blue',label='a')
axes[0].quiver(0,0,b[0],b[1],angles='xy',scale_units='xy',scale=1,color='red',label='b')
axes[0].quiver(0,0,*(a+b),angles='xy',scale_units='xy',scale=1,color='green',label='a+b')
axes[0].set_xlim(-1,5); axes[0].set_ylim(-1,5); axes[0].grid(True); axes[0].legend(); axes[0].set_title('Vector Addition')
# Scalar mult
v = np.array([1, 2])
for s, c in zip([1, 2, -1], ['blue','green','red']):
    axes[1].quiver(0,0,*(s*v),angles='xy',scale_units='xy',scale=1,color=c,label=f'{s}v')
axes[1].set_xlim(-3,3); axes[1].set_ylim(-3,3); axes[1].grid(True); axes[1].legend(); axes[1].set_title('Scalar Multiplication')
plt.tight_layout(); plt.show()
```

## 📝 Summary

Chapter 2 covers the fundamental operations on vectors: addition, subtraction, and scalar multiplication. The dot product is introduced as a key operation that measures similarity between vectors and is the backbone of neural network computations. Vector norms quantify the 'length' or 'magnitude' of a vector — L2 (Euclidean) norm is the standard distance measure, while L1 (Manhattan) norm is used in regularization. Unit vectors normalize a vector to have length 1, preserving direction only.
