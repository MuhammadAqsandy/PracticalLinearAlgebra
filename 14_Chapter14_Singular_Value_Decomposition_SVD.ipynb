## Computing SVD

```python
import numpy as np
A = np.array([[1,2,3],[4,5,6]], dtype=float)
U, s, Vt = np.linalg.svd(A)
print('U shape:', U.shape)   # (2,2) left singular vectors
print('s:', s)               # singular values
print('Vt shape:', Vt.shape)  # (3,3) right singular vectors
# Reconstruct A
S = np.zeros((2,3)); np.fill_diagonal(S, s)
A_recon = U @ S @ Vt
print('Reconstruction error:', np.linalg.norm(A - A_recon))
```

## Singular Values and Matrix Properties

```python
A = np.random.randn(4, 6)
U, s, Vt = np.linalg.svd(A)
print('Singular values:', np.round(s, 4))
print('Rank =', np.sum(s > 1e-10))
print('Frobenius norm from SVD:', round(np.sqrt(np.sum(s**2)), 6))
print('NumPy Frobenius:', round(np.linalg.norm(A, 'fro'), 6))
```

## Low-Rank Approximation

```python
np.random.seed(42)
A = np.random.randn(6, 6)
U, s, Vt = np.linalg.svd(A)
for k in [1, 2, 3, 6]:
    Uk = U[:,:k]; sk = np.diag(s[:k]); Vtk = Vt[:k,:]
    Ak = Uk @ sk @ Vtk
    err = np.linalg.norm(A - Ak, 'fro')
    print(f'Rank-{k} approx error: {err:.4f}')
```

## Image Compression with SVD

```python
from skimage import data
import matplotlib.pyplot as plt
img = data.camera().astype(float)  # 512x512 grayscale
U, s, Vt = np.linalg.svd(img, full_matrices=False)
fig, axes = plt.subplots(1, 4, figsize=(16,4))
axes[0].imshow(img, cmap='gray'); axes[0].set_title('Original'); axes[0].axis('off')
for i, k in enumerate([5, 20, 100]):
    img_k = U[:,:k] @ np.diag(s[:k]) @ Vt[:k,:]
    axes[i+1].imshow(img_k, cmap='gray'); axes[i+1].set_title(f'k={k}'); axes[i+1].axis('off')
plt.tight_layout(); plt.show()
```

## SVD for PCA

```python
np.random.seed(0)
X = np.random.multivariate_normal([0,0], [[3,2],[2,2]], 100)
X_c = X - X.mean(axis=0)  # center
U, s, Vt = np.linalg.svd(X_c, full_matrices=False)
print('Principal components (rows of Vt):\n', np.round(Vt, 4))
print('Explained variance:', np.round(s**2 / np.sum(s**2), 4))
```

## 📝 Summary

Chapter 14 covers Singular Value Decomposition (SVD) — the most general and powerful matrix factorization. SVD decomposes any matrix A = U Σ V^T where U and V are orthogonal matrices and Σ contains non-negative singular values. Unlike eigendecomposition (restricted to square matrices), SVD works for any shape. Key applications: image/data compression via truncated SVD (keeping only k largest singular values gives the best rank-k approximation by Eckart-Young theorem), PCA (singular vectors are principal components), and recommender systems. SVD is numerically the most stable factorization.
