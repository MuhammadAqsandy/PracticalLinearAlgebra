## Linear Transformation - Rotation

```python
import numpy as np
theta = np.pi / 4  # 45 degrees
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
v = np.array([1, 0])
v_rotated = R @ v
print('Original vector:', v)
print('Rotated 45°:', np.round(v_rotated, 4))
```

## Visualizing Transformations

```python
import matplotlib.pyplot as plt
# Grid of points
x = np.linspace(-2, 2, 5)
y = np.linspace(-2, 2, 5)
points = np.array([[xi, yi] for xi in x for yi in y]).T
# Shear transformation
shear = np.array([[1, 0.5],[0, 1]])
transformed = shear @ points
plt.figure(figsize=(10, 4))
plt.subplot(121); plt.scatter(*points, c='blue'); plt.title('Original'); plt.grid(True); plt.axis('equal')
plt.subplot(122); plt.scatter(*transformed, c='red'); plt.title('Shear Transform'); plt.grid(True); plt.axis('equal')
plt.tight_layout(); plt.show()
```

## Covariance Matrix

```python
np.random.seed(42)
data = np.random.multivariate_normal([0, 0], [[2, 1],[1, 1]], 100)
cov = np.cov(data.T)
print('Covariance Matrix:\n', np.round(cov, 3))
print('Variance of X:', round(cov[0,0], 3))
print('Variance of Y:', round(cov[1,1], 3))
print('Covariance XY:', round(cov[0,1], 3))
```

## Matrix as Dataset

```python
# Each row = one sample, each column = one feature
data = np.array([[170, 65, 25],   # height, weight, age
                 [165, 55, 30],
                 [180, 75, 22],
                 [155, 50, 28]])
cols = ['Height', 'Weight', 'Age']
print('Dataset shape:', data.shape)
print('Mean per feature:', np.mean(data, axis=0))
print('Std per feature:', np.round(np.std(data, axis=0), 2))
```

## Solving Linear System

```python
# 2x + y = 5
# x + 3y = 10
A = np.array([[2, 1],[1, 3]], dtype=float)
b = np.array([5, 10], dtype=float)
x = np.linalg.solve(A, b)
print('Solution: x =', round(x[0],4), ', y =', round(x[1],4))
print('Verification A @ x =', A @ x)
```
