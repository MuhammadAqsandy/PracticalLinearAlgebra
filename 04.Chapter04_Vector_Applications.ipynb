## Pearson Correlation via Vectors

```python
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 5, 4, 5], dtype=float)
# Mean-center
xc = x - np.mean(x)
yc = y - np.mean(y)
# Pearson correlation = cosine similarity of centered vectors
corr = np.dot(xc, yc) / (np.linalg.norm(xc) * np.linalg.norm(yc))
print('Pearson correlation:', round(corr, 4))
print('NumPy check:', round(np.corrcoef(x, y)[0,1], 4))
```

## Cosine Similarity for Text

```python
# Simple bag-of-words vectors
doc1 = np.array([1, 1, 0, 1, 0])  # 'I love linear algebra'
doc2 = np.array([1, 1, 1, 0, 0])  # 'I love machine learning'
doc3 = np.array([0, 0, 1, 0, 1])  # 'deep machine networks'
cos12 = np.dot(doc1, doc2) / (np.linalg.norm(doc1) * np.linalg.norm(doc2))
cos13 = np.dot(doc1, doc3) / (np.linalg.norm(doc1) * np.linalg.norm(doc3))
print('Similarity doc1-doc2:', round(cos12, 4))
print('Similarity doc1-doc3:', round(cos13, 4))
```

## Mean Centering Data

```python
data = np.array([10, 20, 30, 40, 50], dtype=float)
centered = data - np.mean(data)
print('Original:', data)
print('Centered:', centered)
print('Mean of centered:', np.mean(centered))
```

## Weighted Average

```python
grades = np.array([85, 90, 78, 92])
weights = np.array([0.2, 0.3, 0.2, 0.3])
weighted_avg = np.dot(grades, weights)
print('Grades:', grades)
print('Weights:', weights)
print('Weighted Average:', weighted_avg)
```

## Moving Average (Smoothing)

```python
signal = np.array([1, 3, 2, 5, 4, 6, 5, 7, 6, 8], dtype=float)
window = 3
smoothed = np.convolve(signal, np.ones(window)/window, mode='valid')
plt.figure(figsize=(8,4))
plt.plot(signal, 'o-', label='Original')
plt.plot(range(window-1, len(signal)), smoothed, 's-', label='Moving Avg (w=3)')
plt.legend(); plt.title('Moving Average Smoothing'); plt.grid(True); plt.show()
```

## 📝 Summary

Chapter 4 applies vector concepts to real-world data science problems. Pearson correlation is derived from cosine similarity of mean-centered vectors, connecting statistics to linear algebra. Cosine similarity powers document similarity in NLP and recommendation engines. Mean-centering is a preprocessing step essential for PCA. Weighted averages (dot product of data and weights) are the core of neural network computations. Moving averages demonstrate convolution as a vector operation for time series smoothing.
