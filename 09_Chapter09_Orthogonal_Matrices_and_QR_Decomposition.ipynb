## Orthogonal Matrix Properties

```python
import numpy as np

# Orthogonal matrix: Q.T @ Q = I, det(Q) = ±1
theta = np.pi / 3
Q = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
print('Q.T @ Q:\n', np.round(Q.T @ Q, 10))
print('det(Q):', round(np.linalg.det(Q), 10))
print('Orthogonal:', np.allclose(Q.T @ Q, np.eye(2)))
```

## Gram-Schmidt Orthogonalization

```python
def gram_schmidt(vectors):
    basis = []
    for v in vectors:
        w = v.copy().astype(float)
        for b in basis:
            w -= np.dot(w, b) * b
        if np.linalg.norm(w) > 1e-10:
            basis.append(w / np.linalg.norm(w))
    return np.array(basis)

vectors = [np.array([1,1,0]), np.array([1,0,1]), np.array([0,1,1])]
Q = gram_schmidt(vectors)
print('Orthonormal basis:\n', np.round(Q, 4))
print('Q @ Q.T:\n', np.round(Q @ Q.T, 10))
```

## QR Decomposition

```python
A = np.array([[1,2,3],[4,5,6],[7,8,10]], dtype=float)
Q, R = np.linalg.qr(A)
print('Q (orthogonal):\n', np.round(Q, 4))
print('R (upper triangular):\n', np.round(R, 4))
print('Q @ R = A?', np.allclose(Q @ R, A))
```

## QR for Solving Linear Systems

```python
A = np.array([[1,2],[3,4],[5,6]], dtype=float)
b = np.array([1,2,3], dtype=float)
Q, R = np.linalg.qr(A)
# Least squares via QR
Qtb = Q.T @ b
x = np.linalg.solve(R[:2,:], Qtb[:2])
print('Solution:', np.round(x, 4))
# Verify with lstsq
x_lstsq, _, _, _ = np.linalg.lstsq(A, b, rcond=None)
print('lstsq solution:', np.round(x_lstsq, 4))
```

## Check Orthonormality of Q

```python
Q, R = np.linalg.qr(np.random.randn(4, 4))
print('Q.T @ Q:\n', np.round(Q.T @ Q, 8))
print('Is orthonormal:', np.allclose(Q.T @ Q, np.eye(4)))
```

## 📝 Summary

Chapter 9 explores orthogonal matrices and QR decomposition. Orthogonal matrices preserve lengths and angles (think rotations/reflections) and have the convenient property that their inverse equals their transpose. The Gram-Schmidt process converts any set of linearly independent vectors into an orthonormal basis. QR decomposition factors any matrix into Q (orthogonal) and R (upper triangular), which is numerically stable and used for solving linear systems, computing eigenvalues, and least-squares fitting.
