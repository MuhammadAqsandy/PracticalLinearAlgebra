## Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.linspace(0, 5, 20)
y = 2 + 0.5*x**2 - x + np.random.randn(20)
# Design matrix with polynomial features
A = np.column_stack([np.ones_like(x), x, x**2])
coef, _, _, _ = np.linalg.lstsq(A, y, rcond=None)
print('Coefficients (c0, c1, c2):', np.round(coef, 3))
x_plot = np.linspace(0, 5, 100)
A_plot = np.column_stack([np.ones(100), x_plot, x_plot**2])
y_plot = A_plot @ coef
plt.figure(figsize=(6,4)); plt.scatter(x, y, c='red', label='Data')
plt.plot(x_plot, y_plot, 'b-', label='Polynomial Fit (degree 2)')
plt.legend(); plt.grid(True); plt.title('Polynomial Regression'); plt.show()
```

## Multiple Linear Regression

```python
np.random.seed(1)
X = np.random.randn(50, 3)  # 50 samples, 3 features
true_w = np.array([2.0, -1.0, 0.5])
y = X @ true_w + 0.1 * np.random.randn(50)
A = np.column_stack([np.ones(50), X])
coef, _, _, _ = np.linalg.lstsq(A, y, rcond=None)
print('True weights:', true_w)
print('Estimated weights:', np.round(coef[1:], 4))
```

## Ridge Regression (L2 Regularization)

```python
# x = (A.T A + λI)^-1 A.T b
def ridge_regression(A, b, lam):
    n = A.shape[1]
    return np.linalg.solve(A.T @ A + lam * np.eye(n), A.T @ b)
np.random.seed(2)
A = np.random.randn(10, 5)
b = np.random.randn(10)
for lam in [0, 0.1, 1.0, 10.0]:
    x = ridge_regression(A, b, lam)
    print(f'λ={lam}: ||x||={np.linalg.norm(x):.4f}')
```

## Overfitting Demonstration

```python
np.random.seed(0)
x = np.linspace(0, 1, 10)
y = np.sin(2*np.pi*x) + 0.2*np.random.randn(10)
plt.figure(figsize=(12,4))
for i, deg in enumerate([1, 4, 9]):
    A = np.vander(x, deg+1, increasing=True)
    coef, _, _, _ = np.linalg.lstsq(A, y, rcond=None)
    xp = np.linspace(0, 1, 200)
    Ap = np.vander(xp, deg+1, increasing=True)
    plt.subplot(1,3,i+1); plt.scatter(x, y, c='red')
    plt.plot(xp, Ap@coef, 'b-'); plt.title(f'Degree {deg}'); plt.ylim(-2,2); plt.grid(True)
plt.tight_layout(); plt.show()
```

## 📝 Summary

Chapter 12 applies least squares to practical regression problems. Polynomial regression extends linear regression by adding polynomial features (x², x³) to the design matrix. Multiple regression handles many input features simultaneously. Ridge regression adds an L2 penalty term (λ||w||²) to the cost function, which shrinks weights and prevents overfitting — this is equivalent to adding λI to the normal equations. The overfitting demonstration shows how high-degree polynomials perfectly fit training data but generalize poorly, motivating regularization.
